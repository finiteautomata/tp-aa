```python
%pylab inline
import pandas as pd
import numpy as np
from sklearn.cross_validation import cross_val_score


# Vamos a hacer un concepto simple:
# (X, Y) es 1 si X, Y estÃ¡n en el cuadrado (0.5, 1)

X = np.random.uniform(0, 1, 1000)
Y = np.random.uniform(0, 1, 1000)

df = pd.DataFrame({'X': X, 'Y': Y})

df['spam'] = df.apply(lambda row: row['X'] > 0.5 and row['Y'] > 0.5, axis=1)

df.iloc[0:10]
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB


def predictWith(aNaiveBayesClassifier,df):
    for i in range(1000):
        if df.iloc[i]['spam']:
            print "%s %s" %(df.iloc[i][0] , df.iloc[i][1]) 
            data = df.iloc[i][['X', 'Y']].reshape(1, -1)
            print aNaiveBayesClassifier.predict(data)

def fittearCon(aNaiveBayesClassifier, df):
    aNaiveBayesClassifier.fit(df[['X', 'Y']], df["spam"])
    return aNaiveBayesClassifier


```

```python
def plot_dataset(df):
    ax = df[df.spam == False].plot.scatter(x='X', y='Y', color='Blue')

    df[df.spam == True].plot.scatter(x='X', y='Y', color='red', ax=ax)

plot_dataset(df)
```

```python

unClasificadorBernoulliano = fittearCon(BernoulliNB(),df)

predictWith(unClasificadorBernoulliano,df)
            

```

```python
unClasificadorGaussiano = fittearCon(GaussianNB(),df)

predictWith(unClasificadorGaussiano,df)
            
```

```python

unClasificadorMultinomial = fittearCon(MultinomialNB(),df)

predictWith(unClasificadorMultinomial,df)
            
```
