# TF-IDF

Vamos a hacer Tf-Idf sobre nuestros datos. Primero, boiler-plate canónico.

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Esto agrega al python path el directorio ..
import sys
sys.path.append("..")

import config
from data_builder import load_test_data, load_dev_data, load_small_dev_data


df, target = load_small_dev_data()

best_words = pd.DataFrame()

options = {
    'max_features' : 2000,
    'ngram_range'  : (1, 3),
}
```

# Tf-idf sobre texto plano del mail

Hagamos tf-idf sobre el texto plano del mail a ver qué queda.

```python

transformer = TfidfVectorizer(**options).fit(df.text)

X_train_tf = transformer.transform(df.text)
print X_train_tf.shape
```

```python
best_words["plain_mail"] = transformer.get_feature_names()
```

Hagamos tf-idf acá. Primero, sin stop_words

# Tf-idf sobre el cuerpo del mail

Parseemos primero el cuerpo del mail

```python
print df.payload[:5]
```

```python

transformer = TfidfVectorizer(**options).fit(df.payload)

X_train_tf = transformer.transform(df.payload)

best_words["with_parsing"] = transformer.get_feature_names()
```

# Tf-idf con rango de min_df y max_df

Vemos que hay palabras no muy copadas arriba

```python

transformer = TfidfVectorizer(min_df=0.001, max_df=0.8, **options).fit(df.payload)

X_train_tf = transformer.transform(df.payload)

best_words["with_modified_df"] = transformer.get_feature_names()
```
