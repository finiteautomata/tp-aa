# Tf-idf sobre el cuerpo del mail

Parseemos primero el cuerpo del mail

```python
import email

parser = email.parser.Parser()

df.parsed_emails = df['text'].apply(
    lambda t: parser.parsestr(t.encode('utf-8')))

def get_text_payload(mail):
    payload = mail.get_payload()

    if type(payload) is str:
        return payload
    elif type(payload) is list:
        return ",".join([get_text_payload(m) for m in payload])
    else:
        raise Exception("Tipo de payload ni string ni lista")

df.payload = df.parsed_emails.apply(get_text_payload)
```

```python
print df.payload[:5]
```

```python

transformer = TfidfVectorizer(max_features=500).fit(df.payload)

X_train_tf = transformer.transform(df.payload)

best_words["with_parsing"] = transformer.get_feature_names()
```

# Tf-idf con rango de min_df y max_df

Vemos que hay palabras no muy copadas arriba

```python

transformer = TfidfVectorizer(max_features=500, min_df=0.01, max_df=0.8).fit(df.payload)

X_train_tf = transformer.transform(df.payload)

best_words["with_modified_df"] = transformer.get_feature_names()
```
