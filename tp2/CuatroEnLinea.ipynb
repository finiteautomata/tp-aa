```python
import random


class CuatroEnLinea:
    def __init__(self, player1, player2,cant_filas,cant_columnas):
        
        self.cant_filas = cant_filas
        self.cant_columnas = cant_columnas
        self.player1, self.player2 = player1, player2
        self.player1_turn = random.choice([True, False])
        self.tablero = [[0 for j in range(0,cant_columnas)] for i in range(0,cant_filas)]


    def play_game(self):
        #self.player1.start_game(1)
        #self.player2.start_game(2)
        
        terminamos = False
        
        while not terminamos:
            if self.player1_turn:
                player, symbol = self.player1, 1
            else:
                player, symbol = self.player2, 2

            self.mostrar_tablero()
                        
            columna = player.move(self)-1
            
            self.poner_ficha(columna,symbol)
            
            if self.jugador_gana_con_movimiento(columna):
                print("ganó el jugador " + str(symbol))
                self.mostrar_tablero()
                terminamos = True
           
            
            if self.tablero_lleno():
                print("tablero lleno")
                terminamos = True         

            self.player1_turn = not self.player1_turn

            
            #if self.player_wins():
            #   player.reward(1, self.board)
            #  other_player.reward(-1, self.board)
            # break
            #if self.board_full(): # tie game
            #    player.reward(0.5, self.board)
            #    other_player.reward(0.5, self.board)
            #    break
            #other_player.reward(0, self.board)
            

    def jugador_gana_con_movimiento(self, j):

        fila_disponible = self.fila_disponible_en(j)
        if fila_disponible == -1:
            i = self.cant_filas - 1
        else:
            i = fila_disponible-1
                     
        return  self.four_in_line(i,j,i-3,j) or self.four_in_line(i,j,i-3,j+3) or self.four_in_line(i,j,i,j+3) or self.four_in_line(i,j,i+3,j+3) or self.four_in_line(i,j,i+3,j) or self.four_in_line(i,j,i+3,j-3) or self.four_in_line(i,j,i,j-3) or self.four_in_line(i,j,i-3,j-3)

    def mostrar_tablero(self):
        for i in reversed(range(0,self.cant_filas)):
            print self.tablero[i]
            print "\n"
            
    def jugadasPosibles(self):
        jugadasPosibles = [0]*self.cant_columnas
        for j in range(0,self.cant_columnas):
            for i in range(0,self.cant_filas):
                if self.tablero[i][j] == 0:
                    jugadasPosibles[j] = 1

        return jugadasPosibles

   
    
    def four_in_line(self, i1,j1,i2,j2):

        tablero = self.tablero
        
        if i2 < 0 or i2 >= self.cant_filas or j2 < 0 or j2 >= self.cant_columnas:
            return False
     
        fila = self.build_range(i1,i2)
        columna = self.build_range(j1,j2)
      
        return (tablero[fila[0]][columna[0]] == tablero[fila[1]][columna[1]] == tablero[fila[2]][columna[2]] == tablero[fila[3]][columna[3]]) and tablero[fila[0]][columna[0]] != 0


    
        
    
    def build_range(self,i1,i2):

        rango = []
        if ( i1 < i2 ):
            rango = [i1,i1+1,i1+2,i1+3]
        elif ( i1 == i2 ):
            rango = [i1,i1,i1,i1]
        else:
            rango = [i1,i1-1,i1-2,i1-3]
            
        return rango


    def tablero_lleno(self):
        return not any([columna != 0 for columna in self.jugadasPosibles()])

   
    
    def poner_ficha(self,j,symbol):
        tablero = self.tablero
        columna = j
        for i in range(0,self.cant_filas):
            if tablero[i][columna] == 0:
                tablero[i][columna] = symbol
                break
                
    def fila_disponible_en(self,j):
        for i in range(0,self.cant_filas):
            if self.tablero[i][j] == 0:
                return i
        return -1
                


class Player(object):
    def __init__(self):
        self.breed = "human"

    def move(self, juego):
        return int(raw_input("Qué columna elegís? "))


p1 = Player()
p2 = Player()

t = CuatroEnLinea(p1, p2,6,7)
t.play_game()

```
